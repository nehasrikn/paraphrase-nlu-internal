```python
x = [1, 1, 1, 1, 1, 0]

np.var(x), np.mean(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean = np.mean(x)
low_variance = np.var(x)

# Generate data points from a normal distribution with low variance
data = np.random.normal(np.mean(x), np.var(x), 1000)
data = np.clip(data, 0, 1)  # Clip values outside the range [0, 1]


# Plotting
plt.figure(figsize=(8, 6))
#plt.hist(data, bins=30, density=True, alpha=0.7, color='skyblue')
plt.title('Low Variance Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')


# Plotting the probability density function (PDF) of the normal distribution
x = np.linspace(mean - 3*low_variance, mean + 3*low_variance, 100)
pdf = (1/(np.sqrt(2*np.pi*low_variance**2))) * np.exp(-(x-mean)**2/(2*low_variance**2))
plt.plot(x, pdf, label='PDF', color='orange')
plt.legend()

plt.show()
```
